```python
# Box plot for the effect of fundraising goal on project success
plt.figure(figsize=(12, 8))
sns.boxplot(x='state', y='usd_goal_real', data=df_2018[df_2018['state'].isin(['successful', 'failed'])], palette='coolwarm')
plt.xlabel('Project State')
plt.ylabel('Fundraising Goal in USD')
plt.title('Effect of Fundraising Goal on Project Success (2018 Dataset)')
plt.show()
```
