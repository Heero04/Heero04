## Building a visualization

In this section, we will show you how to build a basic visualization using the Matplotlib library in Python.

```python
# Start by importing the matplotlib library
# As it is so large, we only want the pyplot part
# To save us writing the full library everytime, we will add the alias plt

import matplotlib.pyplot as plt
```

We will start by building a line chart. To do this, we need data for our x & y axis:

```python
# Define the x-axis values
x_values = [1, 2, 3, 4, 5]

# Define the y-axis values
y_values = [10, 8, 6, 4, 2]
```

We can now build our line chart. We will use the plt.subplots() method to create a canvas that we can add our visualisation onto.

```python
fig, ax = plt.subplots()
```

So far we just have a blank set of axes. To build a line chart, we will call the .plot function on the axes (ax) object.

```python
fig, ax = plt.subplots()

ax.plot(x_values, y_values);
```

Finally, we can get rid of the information at the top of our chart by using the function plt.show()

```python
fig, ax = plt.subplots()

ax.plot(x_values, y_values)

plt.show()
```

## Customizing a visualization

Next, let's take a look at how to customize your visualisations.

We can start by changing the line color and adding circles to the chart.

```python
fig, ax = plt.subplots()

ax.plot(x_values, y_values, color = 'red', marker ='o')

plt.show()
```

We can also add titles to our chart and axes and even change the linestyle

```python
fig, ax = plt.subplots()

ax.plot(x_values, y_values, color = 'red', marker ='o', linestyle = 'dashed')
ax.set_xlabel('x_values')
ax.set_ylabel('y_values')
ax.set_title('line chart')

plt.show()
```

We can change the size of our plots by adding a figsize argument to `subplots()`

```python
fig, ax = plt.subplots(figsize = (12,8))

ax.plot(x_values, y_values, color = 'red', marker ='o', linestyle = 'dashed')
ax.set_xlabel('x_values')
ax.set_ylabel('y_values')
ax.set_title('line chart')

plt.show()
```

We can use the subplots method to add other visualisations to our plots:

```python
more_y_values = [5, 8, 1, 7, 12]

fig, ax = plt.subplots(figsize = (12,8))

ax.plot(x_values, y_values, color = 'red', marker ='o', linestyle = 'dashed')
ax.plot(x_values, more_y_values, color = 'blue', marker ='+', linestyle = 'dashdot')
ax.set_xlabel('x_values')
ax.set_ylabel('y_values')
ax.set_title('line chart')

plt.show()
```

Or even make use of the subplots method to create two visualisations on the same canvas:

```python
fig, ax = plt.subplots(figsize = (12,8), ncols = 2)

ax[0].plot(x_values, y_values, color = 'red', marker ='o', linestyle = 'dashed')
ax[1].plot(x_values, more_y_values, color = 'blue', marker ='+', linestyle = 'dashdot')
ax[0].set_xlabel('x_values')
ax[0].set_ylabel('y_values')
ax[0].set_title('line chart')
ax[1].set_xlabel('x_values')
ax[1].set_ylabel('y_values')
ax[1].set_title('line chart')

plt.show()
```

## Visualizing our cars dataset

Now that you have seen a few data viz examples with Matplotlib, let's use data visualization to learn more about our cars dataset.

We'll start by importing and previewing our dataset. 

```python
import pandas as pd
import matplotlib.pyplot as plt
cars_merged = pd.read_csv('Data/cars_merged.csv')
cars_merged.head()
```

For our first visualization, let's create a histogram to see the distribution of annual mileage across all the drivers in our dataset. We'll use the same syntax as before, and add in two vertical lines (using `axvline`) to see the mean and median of our mileage data. By adding in these lines, we can easily check for any skew in our distribution. 

```python
fig, ax = plt.subplots()
ax.hist(cars_merged.Mileage)
plt.show()
```

We'll use the same syntax as before, and add in two vertical lines (using `axvline`) to see the mean and median of our mileage data. By adding in these lines, we can easily check for any skew in our distribution. We can further customize our chart by adding a title and ensuring that we have a legend for our vertical lines.

```python
fig, ax = plt.subplots()
ax.hist(cars_merged.Mileage)
ax.axvline(cars_merged.Mileage.mean(), color = 'red', label = 'Mean')
ax.axvline(cars_merged.Mileage.median(), color = 'yellow', label = 'Median')
ax.set_title("Distribution of Annual Mileage")
ax.legend()
plt.show()
```

We can see that the distribution of our mileage data is fairly uniform and that the mean and median are very close to one another. 

Another way to look at the distribution of a series is to create a box plot. A box plot visualizes the interquartile range of a series and allows us to quickly check our data for outliers.

```python
fig, ax = plt.subplots()
ax.boxplot(cars_merged.Mileage, showmeans = True, vert = False)
ax.set_title("Annual Mileage");
```

In this series (mileage), the min and max values fall within 1.5x the interquartile range, so we do not see any outliers in our dataset. 

---

We can also use visualizations to compare values across different categories in our dataset. To do this, we first need to group by our categories. 

Let's start by exploring how different states compare, which will help us see if there are any differences in how drivers use their cars in different states. We'll create a new dataframe called states_grouped using the grouped_by function.

```python
states_grouped = cars_merged.groupby('State').mean(numeric_only = True)
states_grouped
```

This dataframe includes a lot of information! To help us make sense of it, let's create a bar chart to show how annual mileage compares across the different states.

```python
fig, ax = plt.subplots(figsize = (8, 6))
ax.bar(states_grouped.index, states_grouped.Mileage)
ax.set_title("Average Annual Mileage by State")
ax.set_xlabel("State")
ax.set_ylabel("Mileage")
plt.show()
```
