```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Datafiniti_Vegetarian_and_Vegan_Restaurants.csv')
```

```python
# Set the style of all graphs to seaborn:
sns.set_style("darkgrid")
```

```python
# Number of restaurants in each city (top 10 cities)
plt.figure(figsize=(16,8))
df['city'].value_counts()[:10].sort_values(ascending=True).plot(kind='barh', color='red')
plt.title('Number of restaurants in each city (Top 10)', size=20)
plt.xlabel('Number of restaurants', size=15)
plt.ylabel('Cities', size=16)
plt.xticks(size=12)
plt.yticks(size=12)
plt.show()
```

```python
# Number of restaurants in each province (top 10 provinces)
plt.figure(figsize=(16,8))
df['province'].value_counts()[:10].sort_values(ascending=True).plot(kind='barh', color='black')
plt.title('Number of restaurants in each State (Top 10)', size=20)
plt.xlabel('Number of restaurants', size=15)
plt.ylabel('Provinces', size=15)
plt.xticks(size=12)
plt.yticks(size=12)
plt.show()
```

```python
# Most common cuisines (top 10)
plt.figure(figsize=(16,8))
df['cuisines'].value_counts()[:10].sort_values(ascending=True).plot(kind='barh', color='grey')
plt.title('Most common cuisines (Top 10)', size=20)
plt.xlabel('Number of restaurants', size=15)
plt.ylabel('Cuisines', size=15)
plt.xticks(size=12)
plt.yticks(size=12)
plt.show()
```
